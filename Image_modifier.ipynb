```python
import numpy as np
from random import randint
from matplotlib import pyplot as plt
import cv2


src = cv2.imread(r"C:\Users\PICHAU\Downloads\foto1.jpeg")
img = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
 
height, weight, colors_chanel = img.shape

for x in range(0, height):
    for y in range(0, weight):
        red = img.item(y, x, 0)
        green = img.item(y, x, 1)
        blue = img.item(y, x, 2)
        
        img.itemset((y, x, 0), red + randint(-255, 255))
        img.itemset((y, x, 1), green + randint(-255, 255))
        img.itemset((y, x, 2), blue + randint(-255, 255))
        
```
