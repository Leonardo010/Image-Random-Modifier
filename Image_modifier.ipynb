```python
import numpy as np
from random import randint
from matplotlib import pyplot as plt
import cv2


src = cv2.imread(r"C:\Users\PICHAU\Downloads\foto.jpeg")
img = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)
 
height, weight, colors_chanel = img.shape

red_random_factor = [0, 50]
green_random_factor = [0, 150]
blue_random_factor = [0, 100]

for y in range(0, height):
    for x in range(0, weight):
        red = img.item(y, x, 0)
        green = img.item(y, x, 1)
        blue = img.item(y, x, 2)
        
        img.itemset((y, x, 0), red + randint(red_random_factor[0], red_random_factor[1]))
        img.itemset((y, x, 1), green + randint(green_random_factor[0], green_random_factor[1]))
        img.itemset((y, x, 2), blue + randint(blue_random_factor[0], blue_random_factor[1]))
        
```
